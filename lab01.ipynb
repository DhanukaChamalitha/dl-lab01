```python
exponential_array = np.random.exponential(size=100000)
uniform_array = np.random.rand(100000)
normal_array = np.random.normal(size=100000)
plt.hist(exponential_array, density=True, bins=200, histtype="stepfilled", alpha=0.7, color="green", label="Exponential")
plt.hist(normal_array, density=True, bins=200, histtype="step", color="red", linewidth=2, label="Normal")
plt.hist(uniform_array, density=True, bins=200, histtype="step", color="blue", linewidth=2, label="Uniform")
plt.axis([-2.5, 8, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Comparison of Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("3D Plot of Z = X^2 + Y^2")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
df.head()
```

```python
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
selected_df = df[features]
pearson_corr = selected_df.corr(method='pearson')
spearman_corr = selected_df.corr(method='spearman')
fig, axes = plt.subplots(1, 2, figsize=(18, 8))
sns.heatmap(pearson_corr, annot=True, fmt=".2f", cmap='coolwarm', ax=axes[0])
axes[0].set_title("Pearson Correlation Coefficient")
sns.heatmap(spearman_corr, annot=True, fmt=".2f", cmap='coolwarm', ax=axes[1])
axes[1].set_title("Spearman Rank Correlation Coefficient")
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('Au_nanoparticle_dataset.csv',index_col=0, encoding='latin')
```

```python
columns_to_keep = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
filtered_df = df[columns_to_keep]
```

```python
print(filtered_df.head(20))
```

```python
print(filtered_df.describe())
```

```python
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
for i, feature in enumerate(columns_to_keep):
    ax = axes[i]
    ax.hist(filtered_df[feature], bins=50, color='skyblue', edgecolor='black')
    ax.set_title(f'Histogram of {feature}')
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(filtered_df)
plt.show()
```

```python
g = sns.PairGrid(filtered_df)
g.map_diag(sns.histplot, kde=True)
g.map_lower(sns.kdeplot)
g.map_upper(sns.scatterplot)
plt.show()
```
